```python
#importing libraries

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df =pd.read_csv('https://raw.githubusercontent.com/jatinc2617/Video-Game-Dataset/main/Video_Games.csv')
df.head()
```

```python
df.info()
```

On the Basis of the head and the df.info(), we found that some columns like Year of Release, critic score, user count, critic count are incorrect. They are expressed as float but would be sufficient to just take the integral part. Few NA's are also observed so the data needs to be cleaned based on the NA's count. 



```python
########################## Data Cleaning ###############################
# Checking NA values 
df.isnull().sum()
```

```python
#temp = df.dropna(subset= ["Year_of_Release","Genre"],inplace= False)
df.dropna(inplace= True)
df.reset_index(inplace = True, drop = True)     #Since the NA is very large we droped the rows which had null values in any field. 
df.isnull().sum()
```

```python
df.info()  #Now, we have 6825 rows with no Null values.
```

```python
# Now now we will worn on datatypes. 
#df['Dates'] = pd.to_datetime(df['Dates'], format='%Y%m%d')
df['Year_of_Release']= pd.to_datetime(df['Year_of_Release'], format='%Y-%m-%d').dt.year

df['Critic_Count']= df['Critic_Count'].astype(int)
df['Critic_Score']= df['Critic_Count'].astype(int)
df['User_Count']= df['User_Count'].astype(int)

df['User_Score']= df['User_Score'].astype(float)

df.head()
```
